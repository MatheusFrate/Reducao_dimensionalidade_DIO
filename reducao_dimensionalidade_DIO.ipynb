```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img_exemplo = cv.imread('./imgs/exemplo.png')
img_exemplo = cv.cvtColor(img_exemplo, cv.COLOR_BGR2RGB)
img_exemplo_gray = cv.cvtColor(img_exemplo, cv.COLOR_BGR2GRAY)

plt.imshow(img_exemplo)
plt.title('Imagem original')
plt.show()

plt.imshow(img_exemplo_gray, cmap='gray')
plt.title('Imagem em tons de cinza')
plt.show()

limiar = 235
# img_binaria = cv.threshold(img_exemplo_gray, limiar, 255, cv.THRESH_BINARY)[1]
img_binaria = img_exemplo_gray.copy()
img_binaria[img_binaria > limiar] = 255
img_binaria[img_binaria <= limiar] = 0

plt.imshow(img_binaria, cmap='gray')
plt.title('Imagem binarizada')
plt.show()

suave = cv.GaussianBlur(img_exemplo_gray, (7, 7), 0)
(T, bin) = cv.threshold(suave, 160, 255, cv.THRESH_BINARY)
(T, binI) = cv.threshold(suave, 160, 255, cv.THRESH_BINARY_INV)
resultado = np.vstack([ np.hstack([suave, bin]), np.hstack([binI, cv.bitwise_and(img_exemplo_gray, img_exemplo_gray, mask=binI)]) ])

plt.imshow(resultado, cmap='gray')
plt.title('Imagem binarizada dio')
plt.show()



```
